```python
plt.scatter(bad['Season'],bad['Rating_IMDB'])
plt.title('Rating By Season')
plt.xlabel('Season')
plt.ylabel('IMDB Rating')
plt.show()
```

```python
# Histogram of episode rating using matplotlib
plt.hist(bad['Rating_IMDB'],bins=15,color='skyblue',edgecolor='black')
plt.title('Distribution of Episode Rating')
plt.xlabel('Rating')
plt.ylabel('Episode')
plt.show()

# Box plot of episode ratings using Seaborn
sns.boxplot(x='Season',y='Episode',data=bad)
plt.title('Boxplot of episode rating by season')
plt.show()
```

```python
season_counts = bad['Season'].value_counts()

# Create a bar diagram
plt.bar(season_counts.index, season_counts.values, color='skyblue')
plt.title('Number of Episodes in Each Season of Breaking Bad')
plt.xlabel('Season')
plt.ylabel('Number of Episodes')
plt.show()
```

```python
sns.countplot(x='Rating_IMDB', data=bad)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=bad,x='U.S. viewers_million')
plt.title('Total No.of Viewers')
plt.xlabel('Viewers')
plt.ylabel('Count of viewes')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=bad,x='Duration_mins')
plt.title('Duration')
plt.xlabel('Duration')
plt.ylabel('Count of viewes')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
